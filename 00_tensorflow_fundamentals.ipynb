### Creating random tensors

Random tensors are tensors of some arbitrary size which contain random numbers

```python
# Create two random but the same tensors(using normal distribution)
random_1 = tf.random.Generator.from_seed(7)#set seed for reproducibility
random_1 = random_1.normal(shape=(3,2))
random_2 = tf.random.Generator.from_seed(7)
random_2 = random_2.normal(shape=(3,2))

#Are they equal?
random_1, random_2, random_1 == random_2
```

### Shuffle the order of elements in a tensor

```python
# Shuffle a tensor (valuable for when you want to shuffle your data so the inherent order doesnt effect learning)
not_shuffled = tf.constant([[10,7],[3,4],[2,5]])
# Shuffle our non-shuffled tensor
tf.random.shuffle(not_shuffled)
```

```python
# Shuffle our non-shuffled tensor(global seed and operational level seeds)
tf.random.set_seed(42)
tf.random.shuffle(not_shuffled, seed=42)
```

```python
not_shuffled
```

Must use global as well as operation level seed

```python
tf.random.set_seed(42)#globak level random seed
tf.random.shuffle(not_shuffled, seed=42)#operation level random seed
```

**Matrix Multiplication**
In machine learning amtrix multiplication is one of the most common operations

There are two rules our tensors (or matrices
 need to fulfill if we're going to matrix multiply them:
 1. The inner dimensions must match
 2. The resulting matrix has the shape of the inner dimensions

```python
# Matrix multiplication in tensor flow
print(tensor)
tf.matmul(tensor, tensor)
```

```python
# Matrix multiplication with python operator "@"
tensor @ tensor
```

```python
tensor.shape
```

```python
# Create a tensor (3,2) tensor
X = tf.constant([[1,2],
                 [3,4],
                [5,6]])
Y = tf.constant([[7,8],
                 [9,10],
                [11,12]])
X, Y
```

```python
# Try to matrix multiply tensors of the same shape
X @ tensor
```

```python
# Lets change the shape of y
tf.reshape(Y, shape=(2, 3))
```

```python
# Multiply X by reshape Y
X @ tf.reshape(Y, shape=(2, 3))
```

```python
tf.matmul(X, tf.reshape(Y, shape=(2,3)))
```

```python
tf.matmul(tf.reshape(X, shape=(2, 3)), Y)
```

```python
# Can do the same with transpose
X, tf.transpose(X), tf.reshape(X, shape=(2, 3))
```

```python
# Try matrix multiplication with transpose rather than reshape
tf.matmul(tf.transpose(X), Y)
```

**The dot product**

Matrix multiplication is also referred to as the dot product.


You can perform matrix multiplication using:
* tf.matmul()
* tf.tensordot()

```python
# Perform the dot product on X and Y( reuqires X or Y to be transposed)
tf.tensordot(tf.transpose(X), Y, axes=1)
```

```python
# Perform matrix multiplication between X and Y (transposed)
tf.matmul(X, tf.transpose(Y))
```

```python
# Perform matrix multiplication between X and Y (reshaped)
tf.matmul(X, tf.reshape(Y, shape=(2, 3)))
```

```python
# Check the values of Y, and reshape Y and transposed Y
print("Normal Y:")
print(Y, "\n") # "\n" is for newline

print("Y reshaped to (2, 3):")
print(tf.reshape(Y, (2, 3)), "\n")

print("Y transposed:")
print(tf.transpose(Y))
```

```python
tf.matmul(X, tf.transpose(Y))
```

Generally, when performing matrix multiplication on two tensors and one of the axes doesn't line up, you will transpose (rather than reshape) one of the tensors to get satisfy the matrix multiplication rules

### Aggregating tensors

Aggregating tensors = condensing them from multiple values down to a smaller amount of values

```python
# Get the absolute values
D = tf.constant([-7, -10])
D
```

```python
# Get the absolute values
tf.abs(D)
```

Lets go through the following forms of aggregation
* Get the minimum
* Get the maximum
* Get the mean of a tensor
* Get the sum of a tensor

```python
# Create a random tensor with values between 0 and 100 of size 50
E = tf.constant(np.random.randint(0, 100, size=50))
E
```

```python
tf.size(E), E.shape, E.ndim
```

```python
# Find the minimum 
tf.reduce_min(E)
```

```python
#Find the maximum
tf.reduce_max(E)
```

```python
# Find the mean
tf.reduce_mean(E)
```

```python
# Find the sum
tf.reduce_sum(E)
```

```python
import tensorflow_probability as tfp
```

```python
#Find the variance (need tensor flow probability)
tfp.stats.variance(E)
```

```python
#Find the std deviation (need tensor flow probability)
tf.math.reduce_std(tf.cast(E, dtype=tf.float32))
```

```python
#Find the variance 
tf.math.reduce_variance(tf.cast(E, dtype=tf.float32))
```

### Find the poistional maximum and minimum

```python
# Create a new tensor for finding poitional minimum and maximum
tf.random.set_seed(42)
F = tf.random.uniform(shape=[50])
F
```

```python
# Find the positional maximum
tf.argmax(F)
```

```python
# Index on our largest value position
F[tf.argmax(F)]
```

```python
# Find the max value of F
tf.reduce_max(F)
```

```python
# Check for equality
F[tf.argmax(F)] == tf.reduce_max(F)
```

```python
# Find the positional minimum
tf.argmin(F)
```

```python
# Find the minimum using the positional minimum index
F[tf.argmin(F)]
```
